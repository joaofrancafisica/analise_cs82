```python
import numpy as np

def rosenbrock(x, N):
    print(N)
    out = 0.0
    for i in range(N-1):
        out += 100.0 * (x[i+1] - x[i]**2)**2 + (1 - x[i])**2
    return out

# slow optimize
var_number = 1000
```

```python
from pyswarms.single.global_best import GlobalBestPSO

# instatiate the optimizer
#x_max = 10 * np.ones(2)
#x_min = -1 * x_max
#bounds = (x_min, x_max)
options = {'c1': 0.5, 'c2': 0.3, 'w': 0.9}
optimizer = GlobalBestPSO(n_particles=10, dimensions=var_number, options=options)

# now run the optimization, pass a=1 and b=100 as a tuple assigned to args

cost, pos = optimizer.optimize(rosenbrock, 1000, N=var_number)
```
